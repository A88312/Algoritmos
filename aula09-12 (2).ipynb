## Interface (exercÃ­cio 3)

Especifica uma interface para a tua aplicaÃ§Ã£o onde seja possÃ­vel visualizar as operaÃ§Ãµes possÃ­veis e as Ãºltimas 10 entradas no diÃ¡rio (ordem cronolÃ³gica decrescente).

```python
def calcEntradas(d):
    lista=[]
    for e in d:
        if 'descrição' in e:
            lista.append(e['data'] + ": " + e['descrição'][0:27])
        else:
            lista.append(e['data'] + ": ...")
    return lista

#lentradas=calcEntradas(MyDiary)
#print(lentradas)    debug-> deteção erros

# Interface
import PySimpleGUI as sg
MyDiary = []


menu_list_column=[
    [sg.Button("Limpar BD")],
    [sg.Button ("Carregar BD")],
    [sg.Button ("Gravar BD")],
    [sg.Button ("Inserir")],
    [sg.Button ("Sair")],
]

entradas=calcEntradas(MyDiary)
data_viewer_column=[
    [sg.Listbox(values=entradas, size= (40,20), key="_entradas")]  #40 é o horizontal?
]
interfacePrincipal= [
    [
        sg.Column(menu_list_column),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]

window= sg.Window("O meu diário", font="Helvetiva 24",
                  default_element_size=(20,1)).Layout(interfacePrincipal)

stop= False
while not stop:
    event, values = window.read()
    # End program if user closes window or
    # presses the OK button
    if event == "Sair" or event== sg.WIN_CLOSED:
        stop = True
    elif event == "Limpar BD":
        entradas=[]
        MyDiary.clear()
        window.find_element('_entradas').Update(values=entradas)
    elif event == "Carregar BD":
        MyDiary == carregarBD("diario-atp.json")
        entradas= calcEntradas(MyDiary)
        window.find.element('_entradas').Update(values=entradas)
    elif event== "Gravar BD":
        gravarBD(MyDiary, "diario-atp.json")
window.close()
```
